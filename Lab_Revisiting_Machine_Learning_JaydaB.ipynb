Check for null values in all the columns

```python
nulls = data.isnull().sum()
print(nulls)
```

Exclude the following variables by looking at the definitions. Create a new empty list called drop_list. We will append this list and then drop all the columns in this list later:

OSOURCE - symbol definitions not provided, too many categories
ZIP CODE - we are including state already

```python
excluded_var = ['OSOURCE', 'ZIP CODE'] #vars to exclude
drop_list = [] #create a list
for col in excluded_var: #check columns and defintions
    if col in data.columns:
        drop_list.append(col)
```

```python
data = data.drop(columns=drop_list) #drop the colums in the list
```

Identify columns that over 85% missing values

```python
nulls = pd.DataFrame(data.isna().sum()*100/len(data), columns=['percentage'])
nulls.sort_values('percentage', ascending = False).head(50)
```

Remove those columns from the dataframe

```python
cols_to_keep = list(nulls[nulls['percentage'] < 85].index)
len(cols_to_keep)
```

```python
data[cols_to_keep]
```

Reduce the number of categories in the column GENDER. The column should only have either "M" for males, "F" for females, and "other" for all the rest

```python
gender_value_counts = data['GENDER'].value_counts() #some nulls and randoms appear
print(gender_value_counts)
```

```python
def gender_simplified(value):    #function to categorize gender
    if value not in ['M', 'F']:
        return 'OTHER'
    return value
data['GENDER'] = data['GENDER'].apply(gender_simplified)
gender_new_counts = data['GENDER'].value_counts()
print(gender_new_counts)
```
